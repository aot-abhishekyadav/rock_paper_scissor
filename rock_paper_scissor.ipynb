```python

import random
def play():
  computer = random.choice([-1,0,1])
  print("your choices are:\nwater\ngun\nsnake")
  your = input("Enter your choice : ").lower()
  dict = {"snake" : 1, "gun" : 0 , "water" : -1}
  reversedict = {1 : "snake", 0 : "gun", -1 : "water"}

  you = dict[your]
  print(f"your choice is {reversedict[you]}\ncomputer choice is {reversedict[computer]}")

  if(computer == you):

    print("Draw")

  else :
    if(computer == 1 and you == -1):
        print("You lose")

    elif(computer == 0 and you == -1):
        print("You win")

    elif(computer == 0 and you == 1):
        print("You lose")

    elif(computer == -1 and you == 1):
        print("You win")

    elif(computer == 1 and you == 0):
        print("You win")

    elif(computer == -1 and you == 0):
        print("You lose")

    else :
        print("something went wrong")
play()
play()
play()
```
